```python
#3 

def get_suffix(filename, dot = False):
    pos = filename.rfind('.')
    if pos > 0 and pos <len(filename)-1 :
        result = filename[pos::] if dot else pos +1
        return result
    else:
        return 'no suffix'
    
get_suffix('aaaasd.asdaaaaa',1)
```

```python
# 4 
def mm(a):
    m1,m2 = (a[0],a[1]) if a[0]>a[1] else (a[1],a[0])
    for i in range(2,len(a)):
        if a[i]>m1:
            m2 = m1
            m1 = a[i]
        elif a[i]>m2:
            m2=a[i]
    return m1,m2
    
a=[1,2,3,4,5,6]
mm(a)
```

```python
##5
def isleap(year):
    if year % 4 ==0 and year %400 !=0 or year % 100 ==0:
        return 1
    else:
        return 0
    
def which_day(year,month,day):
    rday = 0
    l_day = [[31,28,31,30,31,30,31,31,30,31,30,31],[31,29,31,30,31,30,31,31,30,31,30,31]][isleap(year)]
    for i in range(month-1):
        rday += l_day[i]
    rday += day
    return rday

which_day(1993,1,1)
        
    
```

```python
# 6 

def pt(n):
    yh = [[]] * n
    for row in range(len(yh)):
        yh[row] = [None]*(row+1)
        for col  in range(len(yh[row])):
            if col==0 or col == row:
                yh[row][col]=1
            else:
                yh[row][col] = yh[row-1][col] + yh[row-1][col-1]
            print(yh[row][col],end='\t')
        print()
        
        
def main():
    n = int(input('='))
    pt(n)

if __name__=='__main__':
    main()
```

```python
def main():
    num = int(input('Number of rows: '))
    yh = [[]] * num
    for row in range(len(yh)):
        yh[row] = [None] * (row + 1)
        for col in range(len(yh[row])):
            if col == 0 or col == row:
                yh[row][col] = 1
            else:
                yh[row][col] = yh[row - 1][col] + yh[row - 1][col - 1]
            print(yh[row][col], end='\t')
        print()


if __name__ == '__main__':
    main()
```
